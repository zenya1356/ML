Задание обработать данные и получить модель для предсказывания цены на дома из датасета цен домов в калифорнии

### Импорт библиотек

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

### Импорт датасета и первоначальный осмотр

```python
from sklearn.datasets import fetch_california_housing

cal = fetch_california_housing(as_frame=True)  # возвращает frame с X и target
df = cal.frame.copy()
df.head()
```

```python
df.info()
```

Пропусков нет, все данные числовые

### Обработка пропусков
(если бы были)

```python
missing_share = df.isna().mean().sort_values(ascending=False)
print("Доля пропусков по столбцам:\n", missing_share)

threshold = 0.30
to_drop = missing_share[missing_share > threshold].index.tolist()
df = df.drop(columns=to_drop)

# Все колонки числовые
df = df.fillna(df.median(numeric_only=True))
```

### Проверка на дубликаты

```python
dup_count = df.duplicated().sum()
print(f"Количество дубликатов: {dup_count}")
df = df.drop_duplicates()
```

Дубликатов данных нет

### Графики зависимости таргета от признаков

```python
target = 'MedHouseVal'  

# Колонки признаков
num_cols = df.select_dtypes(include=[np.number]).columns.drop(target)
n = len(num_cols)

# Окно графиков
fig, axes = plt.subplots(nrows=(n+2)//3, ncols=3, figsize=(15, 4*((n+2)//3)))
axes = axes.ravel()

# Заполнение графиков
for i, col in enumerate(num_cols):
    axes[i].scatter(df[col], df[target], s=8, alpha=0.6)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel(target)
    axes[i].set_title(f'{col} vs {target}')
for j in range(i+1, len(axes)):
    axes[j].axis('off')
plt.tight_layout()
plt.show()
```

```python
# Бининг по трём группам по квантилям таргета (он численный, переведем таким образом в категориальный)
df['price_bin'] = pd.qcut(df[target], q=3, labels=['low','mid','high'])

num_cols = df.select_dtypes(include=[np.number]).columns.drop(target, errors = 'ignore')

# Построение графиков распределения
for _ in num_cols:
    sns.barplot(data=df, x='price_bin', y=_, estimator=np.mean, errorbar='sd')
    plt.title(f'Mean {_} by price_bin')
    plt.show()

# Построение boxplot-ов
for _ in num_cols:
    sns.boxplot(data=df, x='price_bin', y = _)
    plt.title(f'Boxplot of {_} by price_bin')
    plt.show()

```

### Корреляционный анал

```python
corr = df.select_dtypes(include=[np.number]).corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title("Корреляционная матрица")
plt.show()

corr[target].sort_values(ascending=False)
```

```python
# Удалим таргет и новый категориальный таргет, чтобы не скормить модели как фичи
X = df.drop(columns=[target, 'price_bin'])
y = df[target]

num_cols = X.select_dtypes(include=[np.number]).columns.tolist()

preprocess = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', num_cols)
    ],
    # Чтоб наверняка убрать все лишнее из матрицы признаков модельки
    remainder='drop'
)
```

```python
preprocess.get_feature_names_out
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

### Пайплайн + кросс-валидация + обучение модели

```python

pipe = Pipeline(steps=[
    ('prep', preprocess),
    ('model', DecisionTreeRegressor(random_state=42))
])

param_grid = {
    'model__max_depth': [3, 5, 8, 12, None],
    'model__min_samples_leaf': [1, 5, 10, 20]
}

grid = GridSearchCV(
    pipe, param_grid=param_grid,
    scoring='neg_mean_absolute_error',  # оптимизируем MAE
    cv=5, n_jobs=-1
)
grid.fit(X_train, y_train)

print("Лучшие параметры:", grid.best_params_)
best_model = grid.best_estimator_
```

### Метрики

```python
import sklearn
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

y_pred = best_model.predict(X_test)

mae  = mean_absolute_error(y_test, y_pred)
mse  = mean_squared_error(y_test, y_pred)  
rmse = np.sqrt(mse)                        
r2   = r2_score(y_test, y_pred)

print(f"MAE:  {mae:.3f}")
print(f"RMSE: {rmse:.3f}")
print(f"R2:   {r2:.3f}")
```

## Feature importance

```python
prep = best_model.named_steps['prep']
model = best_model.named_steps['model']

feature_names = []
feature_names.extend(num_cols)

importances = pd.Series(model.feature_importances_, index=feature_names).sort_values(ascending=False)
display(importances.head(15))

plt.figure(figsize=(8,6))
importances.head(15).plot(kind='barh')
plt.gca().invert_yaxis()
plt.title("Top-15 Feature Importances (Decision Tree)")
plt.show()
```

## Визуализация результата для заказчика

```python
# График правда-прогноз
plt.figure(figsize=(5,5))
plt.scatter(y_test, y_pred, s=8, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], linestyle='--')
plt.xlabel("Истинные значения")
plt.ylabel("Прогнозы")
plt.title("Predicted vs True")
plt.show()

# Возьмем урезанную версию дерева для наглядности
small_tree = Pipeline(steps=[
    ('prep', preprocess),
    ('model', DecisionTreeRegressor(max_depth=3, random_state=42))
]).fit(X_train, y_train)

# Обьект дерева получившегося
tree_model = small_tree.named_steps['model']

plt.figure(figsize=(16,8))
plot_tree(tree_model,
          feature_names=feature_names,
          filled=True, rounded=True, fontsize=8)
plt.show()
```
