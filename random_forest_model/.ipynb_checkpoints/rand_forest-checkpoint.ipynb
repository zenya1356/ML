```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    roc_auc_score, roc_curve
)
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import PartialDependenceDisplay
```

```python
data = load_breast_cancer()

df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

df.info()
```

```python
display(df.dtypes)

treshold = 0.3
cols_to_drop = list(df.columns[df.isna().mean() > treshold])
print(cols_to_drop)
df.drop(cols_to_drop, axis=1)

num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
for c in num_cols:
    if df[c].isna().any():
        df[c] = df[c].fillna(df[c].median())
```

```python
print(df.duplicated().sum())
```

Дубликатов строк нет, пропусков в данных нет, типы все числовые

### Исследовательский анализ данных

#### Scater

```python
y = df['target'].values
#шум таргета для визуализации
y_jitter = y + np.random.normal(0, 0.04, size=len(y))
for col in df.columns:
    #чтобы не строить таргет от таргета
    if col == 'target': 
        continue
    plt.figure()
    plt.scatter(df[col].values, y_jitter, s=8, alpha=0.6)
    plt.xlabel(col); plt.ylabel("target (jittered)")
    plt.title(f"{col} vs target")
    plt.tight_layout()
    plt.show()
```

#### Баланс классов

```python
plt.figure()
vals, cnts = np.unique(y, return_counts=True)
plt.bar(vals.astype(str), cnts)
plt.xlabel("target (0=malignant, 1=benign)")
plt.ylabel("count")
plt.title("Class distribution")
plt.tight_layout()
plt.show()
```

Разброс небольшой, полезно понимать, чтобы выбрать метрики 

 #### Boxplot

```python
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()

for col in num_cols:
    s = df[col]
    q_hi = s.quantile(0.9999)
    n_clipped = int((s > q_hi).sum())
    df[col] = s.clip(upper=q_hi)

for col in num_cols:
    plt.figure()
    plt.boxplot(df[col].dropna(), vert=True, whis=1.5, showmeans=False)
    plt.title(f"Boxplot: {col}")
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```

Все признаки нормик выглядят, особых выбросов нет, ничего не отсеклось

#### Кор. матрица

```python
numeric_df = df.select_dtypes(include=[np.number]).drop(columns=['target'], errors='ignore')

corr = numeric_df.corr(numeric_only=True)

mask = np.triu(np.ones_like(corr, dtype=bool))  

threshold = 0.5
annot = corr.round(2).astype(str)
annot_mask = corr.abs() < threshold
annot = annot.where(~annot_mask, other="")

plt.figure(figsize=(14, 11))
sns.heatmap(
    corr,
    mask=mask,
    annot=annot, fmt="",
    cmap="coolwarm", center=0, vmin=-1, vmax=1,
    square=True, linewidths=0.4, linecolor="white",
    cbar_kws={"shrink": 0.7, "ticks": np.linspace(-1, 1, 9)}
)
plt.title("Матрица корреляций")
plt.tight_layout()
plt.show()
```

Есть сильно кореллирующие между собой признаки, но для случайного леса это не страшно, оставим

Разделение со стратификацией, тк дисбаланс присутствует

```python
X = df.drop(columns=['target']).values
y = df['target'].values
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

Непосредственно обучение( 300 деревьев без ограничения по глубине, sqrt для отбора подмножества признаков хорошо для классификации)

```python
rf = RandomForestClassifier(
    n_estimators=300, max_depth=None, max_features="sqrt",
    random_state=42, n_jobs=-1
)
rf.fit(X_train, y_train)
```

#### Предсказация и метрики

```python
y_pred = rf.predict(X_test)
y_proba = rf.predict_proba(X_test)[:, 1]

print("Accuracy:", round(accuracy_score(y_test, y_pred), 4))
print("Precision:", round(precision_score(y_test, y_pred), 4))
print("Recall:", round(recall_score(y_test, y_pred), 4))
print("ROC-AUC:", round(roc_auc_score(y_test, y_proba), 4))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=["malignant(0)", "benign(1)"]))
```

#### Feature importances

```python
importances = rf.feature_importances_
feat_names = df.drop(columns=['target']).columns
feat_imp = pd.Series(importances, index=feat_names).sort_values(ascending=False)

plt.figure()
feat_imp.head(20).iloc[::-1].plot(kind='barh')
plt.title("Top 20")
plt.tight_layout()
plt.show()
```
